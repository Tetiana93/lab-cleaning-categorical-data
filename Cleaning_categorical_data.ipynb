```python
for col in categorical_df.columns:
    sns.barplot(x=categorical_df[col], y=customer_df['total_claim_amount'])
    plt.xticks(rotation=90)
    plt.show()
```

We can see correlation between target column - total_claim_amount and the next columns: coverage - premium, employmentstatus(unemployed), gender(m), location_code(Suburban), marital_status(single), vehicle_class(luxury), vihicle_size(small). For all of this columns for mentioned values we can observe higher total_claim_amount. Fow other columns we can not see big difference between features.
